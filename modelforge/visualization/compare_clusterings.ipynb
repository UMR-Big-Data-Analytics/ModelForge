```python
sns.set_style("white")


def hist_distances(x, ax=None):
    dim = x.shape[1]
    max_dist = np.sqrt(dim)
    dist = pdist(x).flatten()
    with plt.style.context("science"):
        if ax is None:
            ax = plt.gca()
        sns.histplot(dist, bins=30, stat="density", ax=ax)
        ax.axvline(max_dist, color='black', linestyle='--')
        ax.set_xlabel("Distance")
```

```python
from scipy.spatial.distance import pdist
from sklearn.preprocessing import MinMaxScaler

embedding_heating = pd.read_csv("../../data/results/embedding_dimensionality/heating/pairwise_loss_2025-02-17-11:13/pairwise_loss_max_target_entropy_10/137/embedding.csv")
embedding_heating = embedding_heating.drop(columns=["Unnamed: 0", "cluster"])
embedding_heating = MinMaxScaler().fit_transform(embedding_heating)

embedding_anomaly = pd.read_csv("../../data/results/embedding_dimensionality/anomaly/pairwise_loss_2025-02-17-11:13/pairwise_loss_max_target_entropy_10/107/embedding.csv")
embedding_anomaly = embedding_anomaly.drop(columns=["Unnamed: 0", "cluster"])
embedding_anomaly = MinMaxScaler().fit_transform(embedding_anomaly)

embedding_housing = pd.read_csv("../../data/results/embedding_dimensionality/house_price/pairwise_loss_2025-02-17-11:13/pairwise_loss_max_target_entropy_10/116/embedding.csv")
embedding_housing = embedding_housing.drop(columns=["Unnamed: 0", "cluster"])
embedding_housing = MinMaxScaler().fit_transform(embedding_housing)

embedding_weather = pd.read_csv("../../data/results/embedding_dimensionality/weather_probabilistic/pairwise_loss_2025-02-17-11:13/pairwise_loss_max_target_entropy_10/24/embedding.csv")
embedding_weather = embedding_weather.drop(columns=["Unnamed: 0", "cluster"])
embedding_weather = MinMaxScaler().fit_transform(embedding_weather)

with plt.style.context("science"):
    fig, axes = plt.subplots(2, 2, figsize=(8, 8/ratio))
    axes = axes.flatten()
    hist_distances(embedding_heating, axes[0])
    axes[0].set_title("Heating")
    hist_distances(embedding_anomaly, axes[1])
    axes[1].set_title("Anomaly")
    hist_distances(embedding_housing, axes[2])
    axes[2].set_title("Housing")
    hist_distances(embedding_weather, axes[3])
    axes[3].set_title("Weather")
    plt.subplots_adjust(wspace=0.2, hspace=0.35)
    plt.savefig("../../reports/figures/embedding_size/datasets_pairwise_distances.png", bbox_inches="tight")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors
import matplotlib.lines as mlines
sns.set_style("white")
# Define embedding sizes to analyze
embedding_sizes = [3, 5, 10, 20]  # Adjust these based on available datasets

colors = sns.color_palette("viridis", 4)
dataset_labels = ["Heating", "Anomaly", "Housing", "Weather"]

# Fixed k for nearest neighbors
k = 5

def get_num_cluster(dataset):
    if dataset == "heating":
        return 137
    elif dataset == "anomaly":
        return 107
    elif dataset == "house_price":
        return 116
    elif dataset == "weather_probabilistic":
        return 24
    else:
        raise ValueError("dataset must be heating, anomaly, or house_price")

def get_file_path(dataset):
    num_clusters = get_num_cluster(dataset)
    return f"../../data/results/embedding_dimensionality/{dataset}/pairwise_loss_2025-02-17-11:13/pairwise_loss_max_target_entropy_{dim}/{num_clusters}/embedding.csv"

distance_results = {}
for dataset in ["heating", "anomaly", "house_price", "weather_probabilistic"]:
    # Placeholder for results
    distance_results_dataset = {}

    for dim in embedding_sizes:
        # Load dataset for the given embedding size
        file_path = get_file_path(dataset)
        embedding_heating = pd.read_csv(file_path).drop(columns=["Unnamed: 0", "cluster"]).values  # Convert to NumPy for processing
        embedding_heating = MinMaxScaler().fit_transform(embedding_heating)

        # Fit Nearest Neighbors set
        neighbors = NearestNeighbors(n_neighbors=k)
        neighbors.fit(embedding_heating)
        distances, _ = neighbors.kneighbors(embedding_heating)

        # Get k-nearest distances
        nearest_distances = np.sort(distances, axis=0)[:, k - 1]

        # Get k-furthest neighbor distances
        furthest_neighbors = NearestNeighbors(n_neighbors=len(embedding_heating))  # Fit all neighbors
        furthest_neighbors.fit(embedding_heating)
        all_distances, _ = furthest_neighbors.kneighbors(embedding_heating)
        furthest_distances = np.sort(all_distances, axis=0)[:, -k] # Take the k-th furthest neighbor

        # Store results
        distance_results_dataset[dim] = (nearest_distances, furthest_distances)
    distance_results[dataset] = distance_results_dataset

with plt.style.context("science"):
    # Plot results
    # 3.5, 2.625
    fig, axes = plt.subplots(2, 2, figsize=(8, 8/ratio))
    axes = axes.flatten()

    for j, dataset in enumerate(["heating", "anomaly", "house_price", "weather_probabilistic"]):
        distance_results_dataset  = distance_results[dataset]
        for i, dim in enumerate(embedding_sizes):
            nearest_distances, furthest_distances = distance_results_dataset[dim]

            # Plot k-nearest distances
            axes[j].plot(np.arange(len(nearest_distances)), nearest_distances, label=f'{dim}D Nearest', color=colors[i])

            # Plot k-furthest distances
            axes[j].plot(np.arange(len(furthest_distances)), furthest_distances, label=f'{dim}D Furthest', color=colors[i], linestyle='dashed')

            #plt.axhline(np.sqrt(dim), color=colors[i], linestyle='dashdot')

        # Custom legend
        nearest_line = mlines.Line2D([], [], color='black', linestyle='solid', label=f'$p={k}$ Nearest Neighbor')
        furthest_line = mlines.Line2D([], [], color='black', linestyle='dashed', label=f'$p={k}$ Furthest Neighbor')
        size_legend = [mlines.Line2D([], [], color=colors[i], label=f's={dim}') for i, dim in enumerate(embedding_sizes)]

        legend1 = plt.legend(handles=size_legend, bbox_to_anchor=(0.45, -0.25), ncol=4, frameon=False)
        legend2 = plt.legend(handles=[nearest_line] + [furthest_line], bbox_to_anchor=(.65, -0.4), ncol=2, frameon=False)
        # Dirty hack for anomaly yticks
        if j == 1:
            axes[j].set_yticks([0,1,2])
        axes[j].set_xlabel('Sorted $p$-distances')
        axes[j].set_ylabel('Distance')
        axes[j].set_title(dataset_labels[j])
        axes[j].yaxis.set_major_formatter(StrMethodFormatter('{x:,.0f}'))
    plt.gca().add_artist(legend1)
    plt.subplots_adjust(wspace=0.2, hspace=0.35)

    plt.savefig("../../reports/figures/embedding_size/knn_graph_dataset.png")
    plt.show()
```
